```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
#plotting fall velocity for droplet 1
plt.title("Graph of fall time to distance travelled for droplet 1")
plt.ylabel("fall time (s)")
plt.xlabel("distance (m)")
y = np.array([41.62,43.03,46.93,37.15,35.03,41.62,38.12,42.78,38.37,40.43,44.87,39.46])
x = np.array([0.0005,0.0005,0.0005,0.0005,0.0005,0.0005,0.0005,0.0005,0.0005,0.0005,0.0005,0.0005])
xerr = 0.00005
yerr = 0.005
plt.errorbar(x, y, xerr, yerr, fmt='o')
def fitfunc(x,m,c):
    return m*x+c
pars, cov = curve_fit(fitfunc,x,y)
plt.plot(x,fitfunc(x,*pars),'g-', linewidth = '2')
```

```python
#plotting rise velocity for droplet 1
plt.title("Graph of rise time to distance travelled for droplet 1")
plt.ylabel("rise time (s)")
plt.xlabel("distance (m)")
y = np.array([6.25,6.93,4.9,5.25,4.25,2.75,2.90,2.56,3.09,2.81,2.56,2.35])
x = np.array([0.0005,0.0005,0.0005,0.0005,0.0005,0.0005,0.0005,0.0005,0.0005,0.0005,0.0005,0.0005])
xerr = 0.00005
yerr = 0.005
plt.errorbar(x, y, xerr, yerr, fmt='o')
def fitfunc(x,m,c):
    return m*x+c
pars, cov = curve_fit(fitfunc,x,y)
plt.plot(x,fitfunc(x,*pars),'g-', linewidth = '2')
```
