```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
#plotting results of 3.1 as asked
plt.title("Graph of y against x for VA = +1000V")
plt.ylabel("y")
plt.xlabel("x")
x = np.array([2,3,4,5,6,7,8,9,10])
y = np.array([.1,.1,.2,.2,.4,.7,1,1.2,1.5])
xerr = 0.1
yerr = 0.1
plt.errorbar(x, y, xerr, yerr, fmt='o')
def fitfunc(x,m,c):
    return m*x**c
pars, cov = curve_fit(fitfunc,x,y)
plt.plot(x,fitfunc(x,*pars),'g-')
```
